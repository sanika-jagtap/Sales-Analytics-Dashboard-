#### Step-by-Step Data Manipulation Code (Sales Dataset)

#### Step 1: Import Pandas & Load Dataset

```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"D:\IAAP\Python\PYTHON practice\sales_dataset_50_records.csv")

# View first 5 rows
df.head()
```

```python
df
```

#### Step 2: Understand Dataset Structure

```python
df.shape #Check rows and columns
```

```python
df.columns #Column names
```

```python
df.info() #Dataset info (data types)
```

#### Step 3: Basic Exploration

```python
df.describe() #Summary statistics
```

```python
df["City"].unique()
df["Product"].unique() #check unique value
```

#### Step 4: Check Missing Values

```python
df.isnull().sum()  #Check Missing Values
```

#### Step 5: Filtering Data

```python
df[df["Total_Sales"] > 50000]   #Orders with sales above 50,000
```

```python
df[df["City"] == "Pune"] #Orders from Pune city
```

```python
df[(df["Product"] == "Laptop") & (df["Discount_%"] > 0)] #Laptop orders with discount
```

#### Step 6: Sorting Data

```python
df.sort_values("Total_Sales", ascending=False)  #Sort by Total Sales (High to Low)
```

```python
df.sort_values("Total_Sales", ascending=False).head(5) #Top 5 highest sales orders
```

#### Step 7: Add New Columns

```python
df["Profit"] = df["Total_Sales"] * 0.20 
df.head()  #Profit column (assume 20% profit margin)
```

```python
df["Sales_With_GST"] = df["Total_Sales"] * 1.18
#Sales after GST (18%)
```

#### Step 8: GroupBy Analysis (Analyst Level)

```python
df.groupby("City")["Total_Sales"].sum() #Total Sales by City
```

```python
df.groupby("Product")["Total_Sales"].mean() #Average sales by Product
```

```python
df.groupby("Product")["Quantity"].sum() #Total Quantity sold per Product
```

#### Step 9: Aggregation with Multiple Columns

```python
df.groupby("City").agg({
    "Total_Sales": "sum",
    "Quantity": "sum",
    "Discount_%": "mean"
})
```

#### Step 10: Rename Columns

```python
df.rename(columns={"Discount_%": "Discount_Percentage"}, inplace=True)
df.head()
```

#### Step 11: Remove Columns

```python
df.drop("Category", axis=1, inplace=True)
df.head()
```

#### Step 12: Remove Duplicate Records

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)  #Remove duplicates
```

#### Step 13: Date Manipulation

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"]) #Convert Order_Date to datetime
```

```python
df["Month"] = df["Order_Date"].dt.month
df.head() #Extract Month
```

#### Step 14: Monthly Sales Trend

```python
monthly_sales = df.groupby("Month")["Total_Sales"].sum()
monthly_sales
```

#### Step 15: Export Clean Dataset

```python
df.to_csv(r"D:\IAAP\Python\PYTHON practice\sales_cleaned_dataset.csv", index=False)
print("Dataset saved successfully!") #Save updated dataset
```

#### VISUALIZATIONS 

#### Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Step 2: Basic Chart Setup

```python
plt.figure(figsize=(8,5))
```

#### Visualization 1: Total Sales by Product (Bar Chart)

```python
#Question: Which product gives highest sales?
```

```python
#Question: Which product gives highest sales?
plt.figure(figsize=(8,5))

df.groupby("Product")["Total_Sales"].sum().plot(kind="bar")


plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

#### Visualization 2: Sales by City (Bar Chart)

```python
#Question: Which city has maximum revenue?
```

```python
#Question: Which city has maximum revenue?
plt.figure(figsize=(8,5))

df.groupby("City")["Total_Sales"].sum().plot(kind="bar")

plt.title("Total Sales by City")
plt.xlabel("City")
plt.ylabel("Revenue")
plt.show()
```

#### Visualization 3: Product Distribution (Pie Chart)

```python
#Question: Which product is ordered most?
```

```python
#Question: Which product is ordered most?
plt.figure(figsize=(7,7))

df["Product"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("Product Order Distribution")
plt.ylabel("")
plt.show()
```

#### Visualization 4: Sales Trend Over Time (Line Chart)

```python
#Question: How sales change day by day?
```

```python
#Question: How sales change day by day?
df["Order_Date"] = pd.to_datetime(df["Order_Date"])

daily_sales = df.groupby("Order_Date")["Total_Sales"].sum()

plt.figure(figsize=(10,5))
plt.plot(daily_sales)

plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

#### Visualization 5: Quantity Sold by Product (Bar Chart)

```python
plt.figure(figsize=(8,5))

df.groupby("Product")["Quantity"].sum().plot(kind="bar")

plt.title("Total Quantity Sold by Product")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()
```

#### Visualization 6: Correlation Heatmap

```python
#Question: Relationship between Quantity, Price, Discount, Sales
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(df[["Quantity", "Unit_Price", "Discount_Percentage", "Total_Sales"]].corr(),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

#### Visualization 7: Sales Distribution (Histogram)

```python
plt.figure(figsize=(8,5))

plt.hist(df["Total_Sales"], bins=10)

plt.title("Sales Distribution")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.show()
```

#### Visualization 8: Box Plot (Outlier Detection)

```python
#Question: Which product has highest sales variation?
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="Product", y="Total_Sales", data=df)

plt.title("Sales Spread by Product")
plt.show()
```

#### Visualization 9: Sales Dashboard Style (Seaborn)

```python
plt.figure(figsize=(10,5))

sns.barplot(x="Product", y="Total_Sales", data=df)

plt.title("Product vs Total Sales")
plt.show()
```
